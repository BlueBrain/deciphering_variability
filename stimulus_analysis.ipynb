## Analysis of reliability during thalamic stimulation
Input: Experiment 25, 30 classes, gid 0 (no grouping)
Warning: loading the correlations takes a while


```python
import bluepy
import numpy as np
import analysis_exp_25_decoupled
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['pdf.fonttype'] = 42
```

```python
import gc

index_correlation = 1
corrs, bins = analysis_exp_25_decoupled.get_correlations(dt=10.0, decouple=False)
print corrs.shape
means_nw = corrs[:, :, index_correlation, :].mean(axis=(-1))
mean_nw = corrs[:, :, index_correlation, :].mean(axis=(0, -1))
errs_nw = corrs[:, :, index_correlation, :].std(axis=-1)/np.sqrt(435)
err_nw = corrs[:, :, index_correlation, :].mean(axis=0).std(axis=-1)/np.sqrt(435)

corrs = 0
gc.collect()
print "network loaded"
```

```python
corrs, bins = analysis_exp_25_decoupled.get_correlations(dt=10.0, decouple=True)
print corrs.shape
means_dc = corrs[:, :, index_correlation, :].mean(axis=(-1))
mean_dc = corrs[:, :, index_correlation, :].mean(axis=(0, -1))
errs_dc = corrs[:, :, index_correlation, :].std(axis=-1)/np.sqrt(300)
err_dc = corrs[:, :, index_correlation, :].mean(axis=0).std(axis=-1)/np.sqrt(300)

corrs = 0
gc.collect()
print "decoupled loaded"
```

```python
print errs_dc.shape
print errs_nw.shape
```

```python
%matplotlib inline
t_start = 50
t_end = 550
fig, axs = plt.subplots(2, figsize=(15, 3))
for i, (mean, err) in enumerate(zip([mean_nw, mean_dc], [err_nw, err_dc])):

    axs[0].errorbar(bins[t_start:t_end], mean[t_start:t_end], yerr=err[t_start:t_end], linewidth=0.8, color=['red', 'black'][i])


axs[1].plot(bins[t_start:t_end], (mean_dc-mean_nw)[t_start:t_end], linewidth=0.8, color='black')
axs[0].set_xlabel('t (ms)')
axs[1].set_xlabel('t (ms)')
axs[0].set_ylabel('r')
axs[1].set_ylabel('dr')
axs[0].set_xlim([1500, 6500])
axs[1].set_xlim([1500, 6500])
axs[0].set_ylim([0, 0.8])
axs[1].set_ylim([0, 0.8])
axs[0].set_yticks([0, 0.2, 0.4, 0.6, 0.8])
axs[1].set_yticks([0, 0.2, 0.4, 0.6, 0.8])

plt.tight_layout()
plt.savefig('figures_stim/correlation.pdf')
```

```python
import bluepy
from bluepy.v2 import Cell

circuit = bluepy.Simulation(analysis_exp_25_decoupled.get_config_pairs_network()[0][0]).circuit
cells = circuit.v2.cells({Cell.HYPERCOLUMN: 2})
gids_all = np.array(cells.axes[0])
    
print gids_all.min()
```

```python
%matplotlib inline
t_start = 520-100
t_end = 540-100
gids_to_plot = np.array([74419, 75270, 81442, 93272])
fig, axs = plt.subplots(3, 4, figsize=(12, 5))
for j, gid in enumerate(gids_to_plot):
    gidx = gid - gids_all.min()
    for i, (mean, err) in enumerate(zip([means_nw[gidx], means_dc[gidx]], [errs_nw[gidx], errs_dc[gidx]])):
        axs[0, j].fill_between(bins[t_start:t_end] +5, mean[t_start:t_end] - err[t_start:t_end],
                                       mean[t_start:t_end] + err[t_start:t_end],
                                        linewidth=0.8,  color=['red', 'black'][i], alpha=0.3)

        axs[0, j].plot(bins[t_start:t_end]+ 5, mean[t_start:t_end], linewidth=0.8, color=['red', 'black'][i],
                      label=['network', 'decoupled'][i])
        axs[0, j].set_xlim([5200, 5400])
        axs[0, j].set_ylabel('r')
        axs[0, j].set_xlabel('t (ms)')

        axs[0, j].set_ylim([-0.1, 0.9])
plt.tight_layout()
axs[0, 0].legend(frameon=False, loc='upper left')
plt.savefig('figures_stim/vm_corr_example.pdf')
```
