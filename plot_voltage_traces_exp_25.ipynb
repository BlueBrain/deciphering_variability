```python
import bluepy
import numpy as np
import analysis_exp_25_decoupled
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['pdf.fonttype'] = 42
%matplotlib inline
```

```python
bcs = analysis_exp_25_decoupled.get_configs_network();
bcs_2 = analysis_exp_25_decoupled.get_configs_decoupled()[0];
print bcs_2
```

```python
gids = [74419, 75270, 81442, 93272]
soma_voltages = []
for j in range(6):
    sim = bluepy.Simulation(bcs[j]);
    soma_df = sim.v2.reports['soma'].data(gids=gids, t_start=5200, t_end=5400)
    soma_voltages.append([soma_df])
    for k in range(5):
        sim = bluepy.Simulation(bcs_2[k]);
        soma_df = sim.v2.reports['soma'].data(gids=gids, t_start=5200, t_end=5400)
        soma_voltages[j].append(soma_df)
;
```

```python
gids = [74419, 75270, 81442, 93272]
soma_voltages_spont = []
for j in range(6):
    sim = bluepy.Simulation(bcs[j]);
    soma_df = sim.v2.reports['soma'].data(gids=gids, t_start=1500, t_end=2000)
    soma_voltages_spont.append([soma_df])
    for k in range(5):
        sim = bluepy.Simulation(bcs_2[k]);
        soma_df = sim.v2.reports['soma'].data(gids=gids, t_start=1500, t_end=2000)
        soma_voltages_spont[j].append(soma_df)
;
```

```python
fig, axs = plt.subplots(4, 4, figsize=(12, 6))

colors = ['#67000d', '#fc9272','#fb6a4a','#ef3b2c','#cb181d','#a50f15']
colors_dec = ['#67000d', '#99d8c9','#66c2a4','#41ae76','#238b45','#006d2c']
line_styles = ['--', '-', '-', '-', '-', '-']
for j in range(5, -1, -1):
    soma_df = soma_voltages_spont[j][0]
    for i, gid in enumerate(gids):
        axs[0, i].plot(soma_df.columns.values/1000.0, soma_df.loc[gid], color=colors[j], linestyle=line_styles[j])
        axs[0, i].set_ylim([-75, -45])
for j in range(5, -1, -1):
    soma_df = soma_voltages_spont[0][j]       
    for i, gid in enumerate(gids):
        axs[1, i].plot(soma_df.columns.values/1000.0, soma_df.loc[gid], color=colors_dec[j], linestyle=line_styles[j])
        axs[1, i].set_ylim([-75, -45])
for j in range(5, -1, -1):
    soma_df = soma_voltages[j][0]
    for i, gid in enumerate(gids):
        axs[2, i].plot(soma_df.columns.values/1000.0, soma_df.loc[gid], color=colors[j], linestyle=line_styles[j])
        axs[2, i].set_ylim([-75, -45])
for j in range(5, -1, -1):
    soma_df = soma_voltages[0][j]       
    for i, gid in enumerate(gids):
        axs[3, i].plot(soma_df.columns.values/1000.0, soma_df.loc[gid], color=colors_dec[j], linestyle=line_styles[j])
        axs[3, i].set_ylim([-75, -45])
for ax in axs[-1, :]:
    ax.set_xlabel('t (ms)')
for ax in axs[:2, :].flatten():
    ax.set_xticks(range(1500, 2100, 100))
for i, ax in enumerate(axs[0, :]):
    ax.set_title(gids[i])
for ax in axs.flatten():
    ax.set_ylabel('V (mV)')
plt.tight_layout()
plt.savefig('figures_stim/voltages.pdf')
```
